```
import os
import numpy as np
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('stopwords')
```

```
dir = './dataset'
txts = os.listdir(dir)
txts.sort()

docs = []
for each in txts:
    f = open(os.path.join(dir, each), 'r')
    docs.append(f.read())

stop_words = nltk.corpus.stopwords.words("portuguese")
```

```
vectorizer = TfidfVectorizer(stop_words=stop_words)
tfidf = vectorizer.fit_transform(docs)

top = 5
tfidf_top = sorted(list(zip(vectorizer.get_feature_names(), tfidf.sum(0).getA1())), key=lambda x: x[1], reverse=True)[:top]

df = pd.DataFrame(tfidf.toarray(), columns=vectorizer.get_feature_names())
```
